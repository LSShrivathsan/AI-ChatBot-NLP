```python
!pip install transformers

import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AdamW
from sklearn.model_selection import train_test_split

texts = [
    "I want to order a new Phone.",
    "What are the latest gadgets available for purchase?",
    "order a smartwatch ",
    "I need to order a new laptop for work.",
    "Which devices do you have in stock?",
    "Where can I purchase a gaming console?",
    "i want a phone",
    "all available devices,available products,in stock products",
    "all gadgets available here",
    "all the devices available",
    "latest smartphone,latest mobile",
    "buy a phone",
    "get a phone",

]
labels = [
    "order_phone",
    "latest_gadgets",
    "order_smartwatch",
    "order_laptop",
    "available_devices",
    "purchase_console",
    "order_phone",
    "available_devices",
    "available_devices",
    "available_devices",
    "latest_gadgets",
    "order_phone",
    "order_phone",

]

model_name = "bert-base-uncased"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(set(labels)))

tokenized_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

input_ids = tokenized_texts["input_ids"]
attention_mask = tokenized_texts["attention_mask"]
labels = torch.tensor([labels.index(label) for label in labels])

train_inputs, val_inputs, train_labels, val_labels = train_test_split(
    input_ids, labels, random_state=42, test_size=0.2
)

train_masks, val_masks = train_test_split(
    attention_mask, random_state=42, test_size=0.2
)

optimizer = AdamW(model.parameters(), lr=1e-5)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

train_data = list(zip(train_inputs, train_masks, train_labels))
val_data = zip(val_inputs, val_masks, val_labels)

epochs = 5

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for inputs, masks, labels in train_data:
        inputs = inputs.unsqueeze(0).to(device)
        masks = masks.unsqueeze(0).to(device)
        labels = labels.unsqueeze(0).to(device)

        optimizer.zero_grad()
        outputs = model(inputs, attention_mask=masks, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        optimizer.step()

    average_loss = total_loss / len(train_data)
    print(f"Epoch {epoch+1}: Average Loss = {average_loss:.4f}")

model.save_pretrained("intent_classification_model")
tokenizer.save_pretrained("intent_classification_model")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import spacy

tokenizer = AutoTokenizer.from_pretrained("intent_classification_model")
model = AutoModelForSequenceClassification.from_pretrained("intent_classification_model")

intent_labels = [
    "order_phone",
    "latest_gadgets",
    "order_smartwatch",
    "order_laptop",
    "available_devices",
    "purchase_console",
]

intent_names = [
    "Order Phone",
    "Latest Gadgets",
    "Order Smartwatch",
    "Order Laptop",
    "Available Devices",
    "Purchase Console",
]

def order_phone():
    print("Ordering your phone, boss...")

def latest_gadgets():
    print("The latest phones and gadgets available for purchase are:")

def order_smartwatch():
    print("Ordering a smartwatch...")

def order_laptop():
    print("Ordering a new laptop for work...")

def available_devices():
    print("The available devices in stock are...")

def purchase_console():
    print("Where would you likely to purchase a gaming console?")

nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
    return entities

user_input = input("Enter: ")

input_ids = tokenizer.encode(user_input, padding=True, truncation=True, return_tensors="pt")

outputs = model(input_ids)
intent_scores = torch.softmax(outputs.logits, dim=1).squeeze().tolist()

entities = extract_entities(user_input)
print("Entities:", entities)

for label, score in zip(intent_labels, intent_scores):
    print(f"Intent: {label}, Relevance Score: {score:.2f}")
predicted_label = torch.argmax(outputs.logits, dim=1).item()

if predicted_label < len(intent_labels):
    intent = intent_labels[predicted_label]
    intent_name = intent_names[predicted_label]
    print("Final Intent:", intent_name)
    top_scores = sorted(intent_scores, reverse=True)[:2]
    score_diff = top_scores[0] - top_scores[1]

    if score_diff < 0.:
        print("The predicted intents are very close")
    else:
        if intent == "order_phone":
            order_phone()
        elif intent == "latest_gadgets":
            latest_gadgets()
        elif intent == "order_smartwatch":
            order_smartwatch()
        elif intent == "order_laptop":
            order_laptop()
        elif intent == "available_devices":
            available_devices()
        elif intent == "purchase_console":
            purchase_console()
        else:
            print("Unknown intent.")
else:
    print("Unknown intent label.")
```
